```python
# =========================
# IMPORTACI√ìN DE LIBRER√çAS
# =========================
import numpy as np
import pandas as pd


# Imputaci√≥n de nulos usando m√©todos avanzados estad√≠sticos
# -----------------------------------------------------------------------
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer

# Librer√≠as de visualizaci√≥n
# -----------------------------------------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt
# Configuraci√≥n

pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python

# =========================
# FUNCI√ìN DE EDA B√ÅSICO
# =========================

""" Funci√≥n para realizar un an√°lisis exploratorio b√°sico (EDA)
    que permite entender la estructura, calidad y contenido del DataFrame."""
    

 # --- Vista inicial de los datos ---
 
def eda_basico(df_hr):
    
    print("üîç PRIMER VISTAZO A LOS DATOS")
    # Muestra las primeras filas del DataFrame.
    print("üìå Primeras filas:")
    display(df_hr.head())
    # Muestra las √∫ltimas filas.
    print("üìå √öltimas filas:")
    display(df_hr.tail())
    # Muestra una muestra aleatoria para detectar patrones raros.
    print("üìå Muestra aleatoria:")
    display(df_hr.sample(5))

    print("üß± ESTRUCTURA DEL DATAFRAME")
    
    print(f"üìê Dimensiones (filas, columnas): {df_hr.shape}\n")
    # Nombres de las columnas.
    print("üìã Columnas:")
    display(df_hr.columns)
    # Informaci√≥n general: tipos, nulos y uso de memoria.
    print("üß† Informaci√≥n general:")
    display(df_hr.info())
    # Tipo de dato de cada columna.
    print("üìä Tipos de datos por columna:")
    display(df_hr.dtypes)
    
    print("üìâ ESTAD√çSTICAS DESCRIPTIVAS")
    # Estad√≠sticas de variables num√©ricas 
    print("üìà Variables num√©ricas:")
    display(df_hr.describe().T)
    # Estad√≠sticas de variables categ√≥ricas
    print("üî§ Variables categ√≥ricas:")
    display(df_hr.describe(include='O').T)
   
    print("üî¢ CARDINALIDAD")
    # N√∫mero total de valores √∫nicos en todo el DataFrame.
    print(f"üîπ Total de valores √∫nicos en el DataFrame: {df_hr.nunique().sum()}\n")
    # N√∫mero de valores √∫nicos por columna.
    print("üîπN¬∫ de valores √∫nicos por columna:")
    display(df_hr.nunique())
    # Valores √∫nicos por columna (todas) .
    print("üîç Valores √∫nicos por columna. nombres valores columnas (todo):")
    for col in df_hr.columns:
        print(f"üìå {col}: {df_hr[col].unique()}")
    # Valores √∫nicos solo para columnas categ√≥ricas.
    print("üîç Valores √∫nicos por columna. nombres valores columnas ( solo categ√≥ricas):")
    for col in df_hr.select_dtypes(include='object').columns:
        print(f"üìå {col}: {df_hr[col].unique()}")
    
    # Conteo de valores nulos por columna.  
    print("\nüö´ VALORES NULOS")
    display(df_hr.isnull().sum())
    # Porcentaje de valores nulos por columna.  
    print("üö´ Porcentaje de valores nulos (%)")
    display(((df_hr.isnull().sum() / df_hr.shape[0]) * 100).round(2))
      
    print("üìé DUPLICADOS")
    # Conteo de filas duplicadas.
    dup_count = df_hr.duplicated().sum()
    print(f"Duplicados: {dup_count}")
       # Muestra duplicados si existen.
    if dup_count > 0:
        print("Los duplicados son:")
        display(df_hr[df_hr.duplicated()])
    else:
        print("‚úÖ No hay filas duplicadas.")
```

```python
# =========================
# CARGA DE DATOS
# =========================

df_hr = pd.read_csv('hr.csv')  
```

```python
eda_basico(df_hr) # Ejecuci√≥n del EDA.
```
